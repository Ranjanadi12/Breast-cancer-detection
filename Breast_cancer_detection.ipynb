```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_train)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier()
```

```python
rf=random_forest.fit(x_train, y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
acc_rf=accuracy_score(y_test,y_pred)
print(accuracy_score(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_score = cross_val_score(random_forest, x,y,cv = 10,scoring = "accuracy")
np.mean(cv_score)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
y_pred_2=knn.predict(x_test)
```

```python
acc_knn=accuracy_score(y_test,y_pred_2)
print(accuracy_score(y_test,y_pred_2))
```

```python
from sklearn.svm import SVC, LinearSVC
svc = SVC()
svc.fit(x_train, y_train)
```

```python
y_pred_3=svc.predict(x_test)
```

```python
acc_svm=accuracy_score(y_test,y_pred_3)
print(accuracy_score(y_test,y_pred_3))
```

```python
linear_svm = LinearSVC()
linear_svm.fit(x_train,y_train)
```

```python
y_pred_4=linear_svm.predict(x_test)
```

```python
acc_linear_svm=accuracy_score(y_test,y_pred_4)
print(accuracy_score(y_test,y_pred_4))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
y_pred_5=dt.predict(x_test)
```

```python
acc_dt=accuracy_score(y_test,y_pred_5)
print(accuracy_score(y_test,y_pred_5))
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
y_pred_6=lr.predict(x_test)
```

```python
acc_lr=accuracy_score(y_test,y_pred_6)
print(accuracy_score(y_test,y_pred_6))
```

```python
Accuracy=pd.DataFrame({'Models':['Random Forest','KNN', 'SVM','Linear SVM','Decision Tree','Logistic regression'],
                      'Accuracy' : [acc_rf,acc_knn,acc_svm,acc_linear_svm,acc_dt,acc_lr]})
```

```python
Accuracy
```

```python
plt.figure(figsize=(10, 6))
plt.bar(Accuracy['Models'], Accuracy['Accuracy'])
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Accuracy of Different Models')
plt.xticks(rotation=45)
plt.show()
```
